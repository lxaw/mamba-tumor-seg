```python
import matplotlib.pyplot as plt

def plot_loss(train_losses, val_losses, model_name):
    epochs = range(1, len(train_losses) + 1)

    plt.plot(epochs, train_losses, 'b-', label='Train Loss')
    plt.plot(epochs, val_losses, 'r-', label='Val Loss')
    plt.title(f'Training and Validation Loss: {model_name}')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()


```

```python
# ResNet18
file_path = '../loss_data/resnet18_loss.txt'
model_name = "ResNet18"


train_losses = []
val_losses = []

with open(file_path, 'r') as file:
    for line in file:
        if line.startswith('Epoch'):
            train_loss = float(line.split(", Train Loss: ")[1].split(",")[0])
            val_loss = float(line.split(", Val Loss: ")[1].split(",")[0])  # Adjusted to properly extract the loss value

            train_losses.append(train_loss)
            val_losses.append(val_loss)

plot_loss(train_losses, val_losses, model_name)

# ResNet50
file_path = '../loss_data/resnet50_loss.txt'
model_name = "ResNet50"


train_losses = []
val_losses = []

with open(file_path, 'r') as file:
    for line in file:
        if line.startswith('Epoch'):
            train_loss = float(line.split(", Train Loss: ")[1].split(",")[0])
            val_loss = float(line.split(", Val Loss: ")[1].split(",")[0])  # Adjusted to properly extract the loss value

            train_losses.append(train_loss)
            val_losses.append(val_loss)

plot_loss(train_losses, val_losses, model_name)

# UNet
file_path = '../loss_data/unet_loss.txt'
model_name = "UNet"


train_losses = []
val_losses = []

with open(file_path, 'r') as file:
    for line in file:
        if line.startswith('Epoch'):
            train_loss = float(line.split(", Train Loss: ")[1].split(",")[0])
            val_loss = float(line.split(", Val Loss: ")[1].split(",")[0])  # Adjusted to properly extract the loss value

            train_losses.append(train_loss)
            val_losses.append(val_loss)

plot_loss(train_losses, val_losses, model_name)

# Regular UMamba
file_path = '../loss_data/reg_umamba_loss.txt'
model_name = "UMamba"


train_losses = []
val_losses = []

with open(file_path, 'r') as file:
    for line in file:
        if line.startswith('Epoch'):
            train_loss = float(line.split(", Train Loss: ")[1].split(",")[0])
            val_loss = float(line.split(", Val Loss: ")[1].split(",")[0])  # Adjusted to properly extract the loss value

            train_losses.append(train_loss)
            val_losses.append(val_loss)

plot_loss(train_losses, val_losses, model_name)


# PP UMamba
file_path = '../loss_data/pp_umamba_loss.txt'
model_name = "PP-UMamba (Ours)"


train_losses = []
val_losses = []

with open(file_path, 'r') as file:
    for line in file:
        if line.startswith('Epoch'):
            train_loss = float(line.split(", Train Loss: ")[1].split(",")[0])
            val_loss = float(line.split(", Val Loss: ")[1].split(",")[0])  # Adjusted to properly extract the loss value

            train_losses.append(train_loss)
            val_losses.append(val_loss)

plot_loss(train_losses, val_losses, model_name)


# Ultra LightM-UNet
file_path = '../loss_data/ultralightmunet_loss.txt'
model_name = "Ultra LightM-UNet"
train_losses = []
val_losses = []

with open(file_path, 'r') as file:
    for line in file:
        if line.startswith('Epoch'):
            train_loss = float(line.split(", Train Loss: ")[1].split(",")[0])
            val_loss = float(line.split(", Val Loss: ")[1].split(",")[0])  # Adjusted to properly extract the loss value

            train_losses.append(train_loss)
            val_losses.append(val_loss)

plot_loss(train_losses, val_losses, model_name)
```
